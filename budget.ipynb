# Budget Tracker
this notebook tracks my monthly spending and earnings, calculates net surplus or deficit per month, and visualizez the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_spending = [18500, 166500, 127900, 139000, 19000, 755000, 21253, 104100, 51500, 57900, 303499, 0]
my_earnings = [0, 41650, 754260, 0, 0, 2976201, 0, 0, 0, 4000, 22000, 66000]
April = ["1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "8th", "9th", "10th", "11th", "12th"]
```

```python
df = pd.DataFrame({
    "April": April,
    "spending": my_spending,
    "earnings": my_earnings
})

df["net"] = df["earnings"] - df["spending"]
df["status"] = df["net"].apply(lambda x: "surplus" if x >= 0 else "deficit")
```

```python
print("total net:", df["net"].sum(), "won")
```

```python
colors = ["green" if x >= 0 else "red" for x in df["net"]]

df.plot(kind="bar", x="April", y=["spending", "earnings"], figsize=(10,6)) 
plt.title("Monthly Budget - Spending vs Earnings")
plt.tight_layout()
plt.show()

df.plot(kind="bar", x="April", y="net", color=colors, figsize=(10,6))
plt.title("Monthly Net - Surplus vs Deficit")
plt.axhline(y=0, color="black", linewidth=0.8)
plt.tight_layout()
plt.show()
```
